```python
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
```

```python
# Define the Component of URL

base_url = "https://www.bigc.co.th/category/"

categories = ["grocery-bakery",
              "eggs-milk-dairy-products",
              "beverages",
              "pantry-and-ingredients",
              "snacks-and-desserts"]

num_pages = 150

session = requests.Session()
```

```python
# Create lists to store the data

product_categories = []
product_titles = []
product_prices = []
```

```python
# Create a Scraping Loop

### Format of URL: https://www.bigc.co.th/category/grocery-bakery?page=2 ###

for x in categories:
    for page in range(1, num_pages +1 , 1):
        url = f"{base_url}{x}?page={page}"
        
        # Send a GET request to the URL
        response = session.get(url)
        print(f"Fetching page: {url}")

        # Check if the request was successful
        if response.status_code == 200:
            
            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(response.content, "html.parser")
            
            # Loop through the product listings and extract the data
            rows = soup.find_all('div', class_='category_result_row__T1F33')
            
            for row in rows:
                cols = row.find_all('div', class_='category_result_col__LmuH8')
                
                for col in cols:
                    products = col.find_all('div', class_='productCard_container__KXMQK')
        
                    for product in products:
                        category = x
                        title = product.find('div', class_ = 'productCard_title__f1ohZ').a.text
                        price = product.find('div', class_='productCard_price__9T3J8').text.strip()
                        price = price.replace('', '').strip()
            
                        product_categories.append(category)
                        product_titles.append(title)
                        product_prices.append(price)
    
        time.sleep(3)
        print("sleep 3 seconds")
```

```python
# Create a DataFrame from the lists and export

bigc = pd.DataFrame({"Product Category": product_categories,
                     "Product Name": product_titles,
                     "Product Price": product_prices})

bigc.to_excel("Data_BigC.xlsx", index = False)
```
